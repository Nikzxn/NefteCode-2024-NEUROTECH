## Создание эмбеддингов

```python
for blend in unique_blends:
  list_of_embeddings = []
  all_smiles = train_smiles_correct[train_smiles_correct['blend_id'] == blend]['smiles']
  for smiles in all_smiles:
    if smiles != 'unknown1':
      # Tokenize the SMILES string --
      inputs = tokenizer(smiles, return_tensors="pt")

      # Получаем все hidden states
      outputs = model(**inputs, output_hidden_states=True)
      hidden_states = outputs.hidden_states

      # Выбираем последний слой для эмбеддинга, или вы можете объединить несколько слоев вместе
      embedding_output = hidden_states[-1]

      # Вы можете усреднить эмбеддинги по всей последовательности, чтобы получить одно векторное представление для всего SMILES
      embeddings = embedding_output.mean(dim=1)

      list_of_embeddings.append(embeddings)

  stacked_embeddings = torch.stack(list_of_embeddings)
  # Вычисляем среднее по первой оси (0 индекс оси)
  mean_embedding = torch.mean(stacked_embeddings, dim=0)
  all_emb.append(mean_embedding.detach().numpy().squeeze(0))
  #train_data_with_embeddings.loc[count, 'emb'] = mean_embedding.detach().numpy().squeeze(0)
  #count += 1
  # Стекаем входные эмбеддинги для создания тензора размера (количество_эмбеддингов x размер_эмбеддинга)
```

```python
import numpy as np
import torch
from torch.utils.data import TensorDataset, DataLoader
from sklearn.model_selection import train_test_split
import torch.nn as nn
import torch.optim as optim
from sklearn.preprocessing import MinMaxScaler

# Нормализация таргетов
#scaler = MinMaxScaler()
#y_scaled = scaler.fit_transform(targets.reshape(-1, 1))  # Обучаем scaler и преобразовываем таргеты

# Предположим, у вас есть массив эмбеддингов X и массив меток y
embe = np.array(all_emb)  # Ваши эмбеддинги
igriki = np.array(targets)  # Ваши таргеты

# Разделяем данные на обучающую и тестовую выборки, преобразуем их в тензоры PyTorch
X_train, X_test, y_train, y_test = train_test_split(embe, igriki, test_size=0.2, random_state=42)
X_train_tensor = torch.from_numpy(X_train).float()
y_train_tensor = torch.from_numpy(y_train).float()
X_test_tensor = torch.from_numpy(X_test).float()
y_test_tensor = torch.from_numpy(y_test).float()

# Оборачиваем данные в DataLoader
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
test_dataset = TensorDataset(X_test_tensor, y_test_tensor)

train_loader = DataLoader(dataset=train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(dataset=test_dataset, batch_size=32, shuffle=False)

# Определяем модель
class RegressionModel(nn.Module):
    def __init__(self, input_dim):
        super(RegressionModel, self).__init__()
        self.fc1 = nn.Linear(input_dim, 128)
        self.fc2 = nn.Linear(128, 64)
        self.fc3 = nn.Linear(64, 32)
        self.fc4 = nn.Linear(32, 1)
        self.relu = nn.ReLU()

    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        x = self.relu(self.fc3(x))
        output = self.fc4(x)  # Для регрессии не используем активацию на последнем слое
        return output

input_dim = embe.shape[1]
model_final = RegressionModel(input_dim)

# Определяем функцию потерь и оптимизатор
criterion = nn.L1Loss()
optimizer = optim.Adam(model_final.parameters())

# Обучение модели
model_final.train()
for epoch in range(10):
    for batch_idx, (data, target) in enumerate(train_loader):
        # Нулируем градиенты
        optimizer.zero_grad()
        # Прямой проход
        output = model_final(data)
        # Рассчитываем потери
        loss = criterion(output.squeeze(), target)
        # Обратное распространение
        loss.backward()
        # Сделаем шаг оптимизатора
        optimizer.step()

# Оценка модели
model_final.eval() # Переключаемся в режим оценки
test_loss = 0
with torch.no_grad():
    for data, target in test_loader:
        output = model_final(data)
        print(output.squeeze(), target)
        test_loss += criterion(output.squeeze(), target).item()

#test_loss /= len(test_loader.dataset)
print(f'Average test loss: {test_loss}')
```
